```python
import os 

frame_num=20

src = '/work/marioeduardo-a/ftc'
images_src=os.path.join(src,"images")
labels_src=os.path.join(src,"labels")

dst = '/work/marioeduardo-a/ftc-train'+str(frame_num)
images_dst=os.path.join(dst,"images")
labels_dst=os.path.join(dst,"labels")
print(dst)
```

```python
import shutil
import os

# create folders
if not os.path.exists(dst):
    os.makedirs(dst)
    for tt in [images_dst,labels_dst]:
        for ff in ['train','valid','test']:
            folder_fn=os.path.join(tt,ff)
            if not os.path.exists(folder_fn):
                os.makedirs(folder_fn)

    #percentages            
    perc={'train':0.75,'valid':0.15,'test':0.10}
    nums={k: int(perc[k]*frame_num) for k in ['train','valid','test']}
    print(nums)

    #copy images and labels
    for tt in ['images','labels']:
        for ff in ['train','valid','test']:
            src_folder=os.path.join(src,tt,ff)
            dst_folder=os.path.join(dst,tt,ff)
            # print("sf:",src_folder)
            i=1
            for file in sorted(os.listdir(src_folder)):
                # print(file)
                src_file=os.path.join(src_folder,file)
                dst_file=os.path.join(dst_folder,file)
                print("s:",src_file)
                print("d:",dst_file)
                print()
                shutil.copy(src_file,dst_file)
                if i>=nums[ff]:
                    break
                i+=1
            print(i)

        # shutil.copy(source,destination)

# print size of dst directories
for tt in ['images','labels']:
    for ff in ['train','valid','test']:
        # src_folder=os.path.join(src,tt,ff)
        dst_folder=os.path.join(dst,tt,ff)
        print(tt,ff,':',len(os.listdir(dst_folder)))
```
