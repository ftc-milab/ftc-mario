```python
cap = cv2.VideoCapture(video)
with open(changes_file, 'r') as changes:
    lines=changes.readlines()
    for i in range(0,len(lines),2):
        # line1=lines[i]
        # line2=lines[i+1]
        frame=int(lines[i].strip().split(' ')[0])
        print('frame',frame)
        # path=os.path.join(changes_folder)
        # if not os.path.exists(path):
        #     os.makedirs(path)
        
        for j in [-1,0,1]:
            if (frame-j)>=1:
                # print(frame)
                cap.set(cv2.CAP_PROP_POS_FRAMES, max(0,frame-j-1))
                ret, img = cap.read()
                cv2.imwrite(os.path.join(changes_folder,str(frame-j)+".jpeg"), img)
        # print('i',i)
     

cap.release()
```
