```python
import cv2
from pathlib import Path
from tqdm import tqdm
```

```python

# file = Path(r"/content/drive/MyDrive/Colab Notebooks/data/fish_detect/train/train.mp4")
file = Path(r"/home/mario/ftc/FTC-2024-data/Train/train.mp4")
```

```python
capture = cv2.VideoCapture(str(file))
```

```python
total_frames = int(capture.get(cv2.CAP_PROP_FRAME_COUNT))
print(total_frames)
```

```python
save = file.parent.parent/r"images"
print(file.parent.parent)
if save.exists() is False:
  raise FileNotFoundError("Video file not exists!")
```

```python
for f in tqdm(range(1,total_frames+1)):
  _, frame = capture.read()
  add = save/f"{f}.jpeg"
  cv2.imwrite(str(add), frame)
```
