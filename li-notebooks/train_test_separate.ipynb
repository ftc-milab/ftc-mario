```python
import numpy as np
from glob import glob
from pathlib import Path
import shutil
from tqdm import tqdm
```

```python
DATA_DIR = Path("/content/gdrive/MyDrive/Colab Notebooks/data/fish_detect")
if not (DATA_DIR).exists():
  raise FileNotFoundError("Video file not exists!")

img_label = ['images', 'labels']
DATASETS = ['train', 'val']

for il in img_label:
  for ds in DATASETS:
     (DATA_DIR / il / ds).mkdir(parents=True, exist_ok=True)
```

```python
image_paths = np.array(glob(f'{DATA_DIR}/images/*.jpeg'))

print(len(image_paths))

np.random.shuffle(image_paths)

ds_split = np.split(image_paths,indices_or_sections=[int(.8*len(image_paths))]
)

dataset_data = zip(DATASETS, ds_split)

for ds, images in dataset_data:
    for img_path in tqdm(images):
        shutil.move(img_path, f'{DATA_DIR}/images/{ds}/')
```

```python
with (DATA_DIR/'train/train_gt_mot.txt').open(mode="r") as f:
  label = f.readlines()
```

```python
len(glob(f'{DATA_DIR}/images/val/*.jpeg'))
```

```python
WIDTH = 2456
HEIGHT = 2058
for ds in DATASETS:
  for img in tqdm((DATA_DIR/'images'/ds).iterdir()):
    with (DATA_DIR/'labels'/ds/f"{img.stem}.txt").open(mode="w") as label_file:
      for f in label[int(img.stem)*10: (int(img.stem)+1)*10]:
        fish = f.strip().split(" ")
        cat = fish[1]
        xc, yc = (float(fish[2])+50)/WIDTH, (float(fish[3])+50)/HEIGHT
        label_file.write(f"{cat} {xc} {yc} {100/WIDTH} {100/HEIGHT}\n")
```
